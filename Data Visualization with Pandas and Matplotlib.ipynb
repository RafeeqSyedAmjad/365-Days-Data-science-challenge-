```python
df = sns.load_dataset('iris')
df.drop('species',inplace = True,axis = 1)
df
```

```python
df.diff().hist(color = '#000000',figsize = (10,10),alpha = 0.8)
```

```python
#To understand the relationship the sepal_length and petal_length 

df.plot.scatter(x = 'sepal_length', y = 'petal_length' ) ##this scatter plot can be used to understand the pattern inside the graph and used to see the relationship between two different features

#in this plot to point one dot we need to have xco-ordinate and y co-ordinate we can also plot this graph in xyz coordiantes also   
```

```python
# to find out the relationship using even the third co-ordinate we can do that using c parameter
df.plot.scatter(x = 'sepal_length', y = 'petal_length',c = 'sepal_width')
```

```python
# if i want to define the size of the dots i can so that using S variable
df.plot.scatter(x = 'sepal_length', y = 'petal_length',c = 'sepal_width',s = 40)
# c parameter define the colour ranges with respect to the particular column
```

```python
## so in scatter  plot we were getting the circles and in hexbin we will get hexagon
df.plot.hexbin(x = 'sepal_length', y = 'petal_length', gridsize = 8)
```
